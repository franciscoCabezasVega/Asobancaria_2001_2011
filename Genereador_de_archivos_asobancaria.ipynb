```
import os
from datetime import datetime

def pedirNumeroEntero():
 
    correcto = False
    num = 0
    while(not correcto):

        try:
            num = int(input("Seleccione una opción: \n"))
            correcto = True
        except ValueError:
            print('Error, seleccione una opción valida\n')
     
    return num
 
salir = False
opcion = 0
 
while not salir:
 
    print("1. Generar Archivo Asobancario 2001")
    print("2. Generar Archivo Asobancario 2011")
    print("3. Salir\n")
    
    print("+----------------+")
    print("+Elige una opcion+")
    print("+----------------+\n")
 
    opcion = pedirNumeroEntero()
 
    if opcion == 1:

        correcto = False

        fecha = datetime.utcnow()

        file = open("asobancaria-2001.txt", "w")

        eA1 = '01'
        eA2 = str(int(input("Ingrese el nit de la empresa sin guión: \n")))
        eA2 = eA2.zfill(10)
        while (not correcto):
          try:
              eA3 = str(int(input("Ingrese la fecha del recaudo sin espacios en formato AAAAMMDD: \n")))
              datetime.strptime(eA3, '%Y%m%d')
              print("Fecha válida\n")
              correcto = True
          except ValueError:
              print("Fecha inválida\n")
        eA4 = '023'
        eA5 = '00000000830925194'
        eA6 = fecha.strftime('%Y%m%d')
        eA7 = fecha.strftime('%H%M')
        eA8 = 'A'
        eA9 = '01'

        encabezadoArchivo = eA1 + eA2 + eA3 + eA4 + eA5 + eA6 + eA7 + eA8 + eA9
        
        file.write(encabezadoArchivo + os.linesep)
        
        nLotes = int(input("Ingrese el número de lotes en el archivo: \n"))

        #inicio ciclo1
        secuenciaLote = 0
        nRegistros = 0
        totalRecaudado = 0

        for x in range(nLotes):

          eL1 = '05'
          eL2 = str(int(input("Ingrese el Código EAN-13 o NIT sin guión: \n")))
          eL2 = eL2.zfill(13)
          secuenciaLote = secuenciaLote + 1
          eL3 = str(secuenciaLote)
          eL3 = eL3.zfill(4)

          encabezadoLote = eL1 + eL2 + eL3 + '                                                                                                                                               '

          file.write(encabezadoLote + os.linesep)

          nDetalles = int(input("Ingrese el número de detalles en el archivo: \n"))

          #inicio ciclo2
          secuenciaDetalle = 1

          for y in range(nDetalles):
            
            rD1 = '06'
            rD2 = str(int(input("Ingrese el número de la liquidación: \n")))
            rD2 = rD2.zfill(48)
            recaudo = int(input("Ingrese el valor recaudado: \n"))
            rD3 = str(recaudo)
            rD3 = rD3.zfill(12) + '00'
            totalRecaudado = totalRecaudado + recaudo
            rD4 = '01'
            rD5 = '01'
            rD6 = '000000'
            rD7 = '000000'
            rD8 = '007'
            rD9 = '0085'
            secuenciaDetalle = secuenciaDetalle + 1
            rD10 = str(secuenciaDetalle)
            rD10 = rD10.zfill(7)
            rD11 = '000'
            nRegistros = nRegistros + 1
            registroDetalle = rD1 + rD2 + rD3 + rD4 + rD5 + rD6 + rD7 + rD8 + rD9 + rD10 + rD11 + '                                                                 '
            
            file.write(registroDetalle + os.linesep)

            # fin ciclo2
        #fin ciclo1
        totalRegistros = nRegistros

        cL1 = '08'
        cL2 = str(totalRegistros)
        cL2 = cL2.zfill(9)
        cL3 = str(totalRecaudado)
        cL3 = cL3.zfill(16) + '00'
        cL4 = '0001'

        controlLote = cL1 + cL2 + cL3 + cL4 + '                                                                                                                                 '

        file.write(controlLote + os.linesep)

        cA1 = '09'
        cA2 = str(totalRegistros)
        cA2 = cA2.zfill(9)
        cA3 = str(totalRecaudado)
        cA3 = cA3.zfill(16) + '00'

        controlArchivo = cA1 + cA2 + cA3 + '                                                                                                                                     '

        file.write(controlArchivo)

        file.close()

    elif opcion == 2:
        salir = True

    elif opcion == 3:
        salir = True
        
    else:
        print("Introduce un numero entre 1 y 3\n")
 
print("Fin\n")

```
